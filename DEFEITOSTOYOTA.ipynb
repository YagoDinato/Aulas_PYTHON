```python
defeitos = []

while True:
    print("\n" + "="*30)
    print(f"{' SISTEMA DE DEFEITOS TOYOTA ':^30}")
    print("="*30)
    print(" 1 Inserir quantidade de defeitos diários durante uma Quinzena")
    print(" 2 Média")
    print(" 3 Maior / Menor consumo")
    print(" 4 ACIMA MÉDIA")
    print(" 5 Dias aceitáveis (defeitos < 10)")
    print(" 6 Ranking top 5 dias com mais falhas")
    print(" 7 Dias com menor defeito (defeitos < 3)")
    print(" 8 Classificação")
    print(" 9 Sair")
    print("="*30)

    try:
        opcao = int(input("Qual opção você deseja? "))

        if opcao == 1:
            defeitos = []
            for i in range(15):
                insdefeito = int(input(f"Insira os defeitos do dia {i + 1}: "))
                defeitos.append(insdefeito)
            print("Defeitos adicionados!")

        elif opcao == 2:
            if defeitos:
                media = sum(defeitos) / len(defeitos)
                print(f"Média de defeitos quinzenal: {media:.2f} defeitos")
            else:
                print("Nenhum defeito cadastrado.")

        elif opcao == 3:
            if defeitos:
                print(f"A maior quantidade de defeitos foi: {max(defeitos)}")
                print(f"A menor quantidade de defeitos foi: {min(defeitos)}")
            else:
                print("Nenhum defeito foi encontrado.")

        elif opcao == 4:
            if defeitos:
                media = sum(defeitos) / len(defeitos)
                acima = [d for d in defeitos if d > media]
                print(f"Quantidade de dias acima da média ({media:.2f}): {len(acima)}")
                print(f"Valores: {acima}")
            else:
                print("Nenhum dado encontrado.")

        elif opcao == 5:
            aceitaveis = [d for d in defeitos if d < 10]
            if aceitaveis:
                print(f"Dias aceitáveis (abaixo de 10): {aceitaveis}")
            else:
                print("Nenhum dia aceitável ou lista vazia.")

        elif opcao == 6:
            if defeitos:
                ranking = sorted(defeitos, reverse=True)[:5]
                print(f'Ranking top 5 (mais falhas): {ranking}')
            else:
                print("Nenhum dado encontrado.")

        elif opcao == 7:
            menores = [d for d in defeitos if d < 3]
            if menores:
                print(f"Dias com menos de 3 defeitos: {menores}")
            else:
                print("Nenhum dia com menos de 3 defeitos.")

        elif opcao == 8:
            if defeitos:
                for d in defeitos:
                    if d > 15:
                        tipo = "CRÍTICA"
                    elif d > 5:
                        tipo = "ATENÇÃO"
                    else:
                        tipo = "CONTROLADA"
                    print(f"{d} DEFEITOS - {tipo}")
            else:
                print("Nenhum dado para classificar.")

        elif opcao == 9:
            print("Saindo do sistema...")
            break

        else:
            print("Opção inválida!")

    except ValueError:
        print("Erro: Por favor, digite apenas números inteiros!")
```
