```
compra = float (input("Qual foi o valor da sua compra?: "))           #Recebe o valor da compra
porcentagem = float (input("Qual é a porcentagem de desconto?:"))     #Recebe o valor da porcentagem
desconto = porcentagem/100 * compra                                   #Faz o conversão de porcentagem e decimal e multiplica a compra
valorfinal = compra - desconto                                        #Compra - Desconto

print (f"O seu valor de desconto é: {desconto}")                      #Mostra o desconto
print (f"O valor final da sua compra é: {valorfinal}")                #Mostra o valor da compra
```

```
peso = float(input("Qual o seu peso em kilogramas? :"))     #Recebe o peso
altura = float(input("Qual a sua altura em metros? :"))     #Rebe a altura
imc = peso / altura**2                                      #Calcula o IMC

print(f"Seu IMC é: {imc:.2f}")                              #Mostra o IMC com so duas casas depois da virgula
if imc<18.5:                                                #Se
  print("Abaixo do peso")
elif imc>=18.5 and imc<=24.9:
  print("Peso normal")
elif imc>=25 and imc<= 29.9:
  print("Acima do peso")
else :
  print("Obeso")

```

```
salario = float(input("Qual é o seu salário?: "))
percentual_aumento = float(input("Qual é o percentual de aumento?: "))
salario_atual = (percentual_aumento/100 +1) * salario
mensagem_final = salario_atual-salario


if mensagem_final <= 350:
  print (f"Seu salário atual é : {salario_atual:.2f}, Parabéns pelo seu aumento")

elif mensagem_final >= 500:
  print (f"Seu salário atual é : {salario_atual:.2f}, Você teve um excelente aumento")

elif mensagem_final >= 720:
  print (f"Seu salário atual é : {salario_atual:.2f}, Você teve um ótimo aumento")

else:
  print (f"Seu salário atual é : {salario_atual:.2f}, Não pude dizer")


```
