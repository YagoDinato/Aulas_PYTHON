```python
acima = 0
abaixo = 0
notas = []
for i in range(6):
  notas_alunos = float(input("Insira 6 notas: "))
  notas.append(notas_alunos)

media = sum(notas) / len(notas)

for n in notas:
  if n >= media:
    acima += 1
  else:
    abaixo += 1

print(f"Média da turma: {media:.2f}")
print(f"Alunos acima ou na média: {acima}")
print(f"Alunos abaixo da média: {abaixo}")
```

```python
numeros = []
for i in range(6):
  numeros_entrada = int(input("Inserir números: "))
  numeros.append(numeros_entrada)

  maior_valor = max(numeros)
  menor_valor = min(numeros)
  posicao_maior = numeros.index(maior_valor)
  posicao_menor = numeros.index(menor_valor)
print(f"O maior número da lista é : {maior_valor}, e posição dele é {posicao_maior +1}")
print(f"O menor valor é: {menor_valor}, e posição dele é {posicao_menor}")
```

```python
parcialidade = []
pares = 0
impar = 0
for i in range(5):
  dig_numeros = int(input("Insira números aleatorios: "))
  parcialidade.append(dig_numeros)

for n in parcialidade:
  if n % 2 == 0:
    pares += 1
  else:
    impar +=1
print(f"Existem {pares} números pares")
print(f"Existem {impar} números ímpares")
```

```python
lista = []
for i in range(3):
  dig_valor = int(input("Insira um valor inteiro: "))
  lista.append(dig_valor)

alvo = int(input("Digite o valor para buscar: "))
encontrado = False

for i in range(len(lista)):
  if lista[i] == alvo:
     print("Valor encontrado na posição", i)
     encontrado = True
     break
if not encontrado:
  print("Valor não encontrado")

```
