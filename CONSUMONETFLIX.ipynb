```python
tempo = []

while True:
    print("\n" + "="*30)
    print(f"{' SISTEMA DE CONSUMO DE NETFLIX ':^30}")
    print("="*30)
    print(" 1 Inserir consumo diário")
    print(" 2 Média")
    print(" 3 ACIMA MEDIA")
    print(" 4 Maior / Menor consumo")
    print(" 5 Consumos acima de 120 min")
    print(" 6 Ranking top 5 ")
    print(" 7 Dias consumo < 30 min ")
    print(" 8 Classificação ")
    print(" 9 Sair")
    print("="*30)

    try:
        opcao = int(input("Qual opção você deseja? "))

        if opcao == 1:
            for i in range(7):
                insconsumo = float(input(f"Insira um consumo em min do dia {i + 1}: "))
                tempo.append(insconsumo)
            print("Consumos adicionados!")

        elif opcao == 2:
            if len(tempo) > 0:
                media = sum(tempo) / len(tempo)
                print(f"Média de consumo: {media:.2f} min")
            else:
                print("Nenhum consumo cadastrado.")

        elif opcao == 3:
            if len(tempo) > 0:
                media = sum(tempo) / len(tempo)
                acima = 0
                for valor in tempo:
                    if valor > media:
                        acima += 1
                print(f"Seus dias acima da média ({media:.2f}) foram: {acima}")
            else:
                print("Nenhum consumo foi encontrado.")

        elif opcao == 4:
            if len(tempo) > 0:
                print(f"O maior consumo é: {max(tempo)} min")
                print(f"O menor consumo é: {min(tempo)} min")
            else:
                print("Nenhum consumo foi encontrado.")

        elif opcao == 5:
            if len(tempo) > 0:
                print("Dias com mais de 120 min:")
                encontrou = False
                for g in tempo:
                    if g > 120:
                        print(f"- {g} min")
                        encontrou = True
                if not encontrou:
                    print("Nenhum dia acima de 120 min.")
            else:
                print("Nenhum consumo foi encontrado.")

        elif opcao == 6:
            if len(tempo) > 0:
                ranking = sorted(tempo, reverse=True)[:5]
                print(f'Ranking top 5: {ranking}')
            else:
                print("Nenhum consumo foi encontrado.")

        elif opcao == 7:
            if len(tempo) > 0:
                print("Dias com menos de 30 min:")
                for c in tempo:
                    if c < 30:
                        print(f"- {c} min")
            else:
                print("Nenhum consumo cadastrado.")

        elif opcao == 8:
            if len(tempo) > 0:
                for t in tempo:
                    if t < 60:
                        tipo = "BAIXO"
                    elif t <= 120:
                        tipo = "REGULAR"
                    else:
                        tipo = "INTENSO"
                    print(f"{t} min - {tipo}")
            else:
                print("Nenhum consumo para classificar.")

        elif opcao == 9:
            print("Saindo do sistema...")
            break

        else:
            print("Opção inválida!")

    except ValueError:
        print("Por favor, digite apenas números!")
```
