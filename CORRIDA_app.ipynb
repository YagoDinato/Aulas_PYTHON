```python
ganhos = []

while True:
    print("\n" + "="*30)
    print(f"{' SISTEMA DE GANHOS DIÁRIOS':^30}")
    print("="*30)
    print(" 1 Inserir ganho")
    print(" 2 Média")
    print(" 3 ACIMA MEDIA")
    print(" 4 Maior / Menor ganho")
    print(" 5 Porcentagem dias acima")
    print(" 6 Classificação ")
    print(" 7 Ranking top 5 ")
    print(" 8 Dias ganho < 70 ")
    print(" 9 Sair")
    print("="*30)

    try:
        opcao = int(input("Insira a opção que você deseja: "))

        if opcao == 1:
          for i in range(7):
            insganho = float(input("Insira uma ganho: "))
            ganhos.append(insganho)
            print("Ganho adicionada!")

        elif opcao == 2:
            if len(ganhos) > 0:
                media = sum(ganhos) / len(ganhos)
                print(f"Média de ganhos: {media:.2f}")
            else:
                print("Nenhuma ganho cadastrada.")

        elif opcao == 3:
            if len(ganhos) > 0:
                media = sum(ganhos) / len(ganhos)
                acima = 0
                for insganhos in ganhos:
                    if insganhos >= media:
                        acima += 1
                print(f"Dias de ganho acima da média: {acima}")
            else:
                print("Nenhuma ganho cadastrada.")

        elif opcao == 4:
            if len(ganhos) > 0:
                print(f"A maior ganho é: {max(ganhos)}")
                print(f"A menor ganho é: {min(ganhos)}")
            else:
                print("Nenhuma ganho cadastrada.")


        elif opcao == 5:
            if len(ganhos) > 0:
                media = sum(ganhos) / len(ganhos)
                acima = 0
                for n in ganhos:
                    if n > media:
                        acima += 1
                acimapor = (acima / 7) * 100

                print(f"Porcentagem acima da média: {acimapor:.2f}% ")
            else:
                print("Nenhuma venda cadastrada.")

        elif opcao == 6:
          for g in ganhos:
            if g < 100:
              tipo = "RUIM"
            elif g <= 300:
              tipo = "BOM"
            else:
              tipo = "EXCELENTE"
            print(f"R$ {g:.2f} - {tipo}")

        elif opcao == 7:
          ranking = sorted(ganhos, reverse = True)[:5]
          print(f'Ranking top 5 = {ranking}')

        elif opcao == 8:
          for g in ganhos:
            if g < 70:
                print(f"R$ {g:.2f}")
            else:
                print("Opção inválida!")

        elif opcao == 9:
            print("Saindo do sistema...")
            break

        else:
            print("Opção inválida!")

    except ValueError:
        print("Erro: Insira apenas números.")
```
