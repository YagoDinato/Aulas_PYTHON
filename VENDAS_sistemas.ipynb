```python
vendas = []

while True:
    print("\n" + "="*30)
    print(f"{' SISTEMA DE VENDAS':^30}")
    print("="*30)
    print(" 1 Inserir venda")
    print(" 2 Média")
    print(" 3 ACIMA MEDIA")
    print(" 4 Maior / Menor nota")
    print(" 5 Porcentagem dias acima")
    print(" 6 Classificação ")
    print(" 7 Ranking top 5 ")
    print(" 8 Sair")
    print("="*30)

    try:
        opcao = int(input("Insira a opção que você deseja: "))

        if opcao == 1:
          for i in range(7):
            insvenda = float(input("Insira uma venda: "))
            vendas.append(insvenda)
            print("Venda adicionada!")

        elif opcao == 2:
            if len(vendas) > 0:
                media = sum(vendas) / len(vendas)
                print(f"Média de vendas: {media:.2f}")
            else:
                print("Nenhuma venda cadastrada.")

        elif opcao == 3:
            if len(vendas) > 0:
                media = sum(vendas) / len(vendas)
                acima = 0
                for insvenda in vendas:
                    if insvenda >= media:
                        acima += 1
                print(f"Vendas dias de vendas acima da média): {acima}")
            else:
                print("Nenhuma venda cadastrada.")

        elif opcao == 4:
            if len(vendas) > 0:
                print(f"A maior venda é: {max(vendas)}")
                print(f"A menor venda é: {min(vendas)}")
            else:
                print("Nenhuma venda cadastrada.")


        elif opcao == 5:
            if len(vendas) > 0:
                media = sum(vendas) / len(vendas)
                acima = 0
                for n in vendas:
                    if n >= media:
                        acima += 1
                acimapor = (acima / 7) * 100

                print(f"Porcentagem acima da média: {acimapor:.2f}% ")
            else:
                print("Nenhuma venda cadastrada.")

        elif opcao == 6:
          for v in vendas:
            if v < 100:
              tipo = "Baixo"
            elif v <= 300:
              tipo = "Medio"
            else:
              tipo = "Alto"
            print(f"R$ {v:.2f} - {tipo}")

        elif opcao == 7:
          ranking = sorted(vendas, reverse = True)[:5]
          print(f'Ranking top 5 = {ranking}')

        elif opcao == 8:
            print("Saindo do sistema...")
            break

        else:
            print("Opção inválida!")

    except ValueError:
        print("Erro: Insira apenas números.")
```
